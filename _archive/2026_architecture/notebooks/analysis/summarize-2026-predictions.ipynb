```python
# Check for local CSV; if missing, download from Firebase using pyasebase/pyrebase
import os
import pandas as pd
from pybaseball import pitching_stats
import numpy as np

csv_path = "./data/pitching_stats.csv"

if os.path.exists(csv_path):
    print(f"Found {csv_path}, reading into DataFrame.")
    data = pd.read_csv(csv_path)
else:
    print(f"{csv_path} not found — attempting to download using pybaseball.")
    qual = 10
    start_season = 2015
    end_season = 2025
    data = pitching_stats(start_season=start_season, end_season=end_season, qual=qual)
    # data.to_csv('./data/pitching_stats.csv', index=False)
```

```python
cnn_results = pd.read_csv('./results/cnn/2026_war_predictions.csv')
mvt_results= pd.read_csv('./results/v3/2026_war_predictions.csv')
```

```python
data_2025 = data.loc[data['Season']==2025, ['IDfg', 'Name', 'IP', 'G', 'GS']]
results_with_name = data_2025.merge(
    cnn_results, left_on = ['IDfg'], right_on = ['player_id'],
    how = 'right'
).merge(
    mvt_results, left_on = ['IDfg'], right_on = ['player_id'],
    how = 'right'
)[['IDfg', 'Name', 'IP', 'G', 'GS','predicted_IP_2026', 'predicted_G_2026', 'predicted_GS_2026','predicted_war_2026_x', 'predicted_war_2026_y']]

```

```python
rho = 0.5
results_with_name['ensemble_prediction'] = ((1-rho) * results_with_name['predicted_war_2026_x'] + rho* results_with_name['predicted_war_2026_y'])
pd.set_option('display.max_row', 100)
results_with_name[results_with_name['predicted_GS_2026']>10].sort_values('ensemble_prediction', ascending=False).head(100)
```
